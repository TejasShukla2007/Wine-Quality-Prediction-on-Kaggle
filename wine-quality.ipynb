```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
train_df
```

```python
train_df.info()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras import regularizers
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = train_df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol' ]]
```

```python
y = train_df[['quality']]
```

```python
y.describe()
```

```python
y_mod = y-3
```

```python
y_mod.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size = 0.3, random_state = 42 )
```

```python
X_cv
```

```python
y_train_mod = y_train - 3
```

```python
y_train_mod.describe()
```

```python
scale = StandardScaler()
X_norm = scale.fit_transform(X_train)
X_test_norm = scale.transform(X_cv)
```

```python
model = Sequential([
    Dense(units = 120, activation = 'relu', name = "L1", kernel_regularizer = regularizers.l2(0.01) ),
    Dense(units = 100, activation = 'relu', name = "L2", kernel_regularizer = regularizers.l2(0.01)),
    Dense(units = 90, activation = 'relu', name = "L3", kernel_regularizer = regularizers.l2(0.01)),
    Dense(units = 60, activation = 'relu', name = "L4", kernel_regularizer = regularizers.l2(0.01)),
    Dense(units = 6, activation = 'linear', name = "Output")
])
model.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001))
```

```python
model.fit(X_norm, y_train_mod, epochs = 100)
```

```python
test_prediction = model.predict(X_test_norm)
```

```python
test_prediction
```

```python
test_pred_class = np.argmax(test_prediction, axis=1)+3
```

```python
test_pred_class
```

```python
train_prediction = model.predict(X_norm)
```

```python
train_pred_class = np.argmax(train_prediction, axis = 1)+3
train_pred_class
```

```python
import math
from sklearn.metrics import accuracy_score
```

```python
print("Training accuracy ", accuracy_score(train_pred_class, y_train))
print("Cross-validation accuracy ", accuracy_score(test_pred_class, y_cv))
```

```python
X_norm_full = scale.transform(X)
X_norm_full
```

```python
model.fit(X_norm_full, y_mod, epochs = 100)
```

```python
pred_full = model.predict(X_norm_full)
```

```python
pred_full_class = np.argmax(pred_full, axis = 1) + 3
pred_full_class
```

```python
print("Accuracy score on training on full data: ", accuracy_score(pred_full_class, y))
```
